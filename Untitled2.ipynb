#read all the images and converting them to array

```python
training_data = []

def creat_training_data ():
    for category in classes:
        path = os.path.join(DataDirectory, category)
        class_num = classes.index(category) #label
        LIMIT_PER_CLASS = 500
        images_loaded = 0
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array= cv2.resize(img_array, (img_size,img_size))
                training_data.append([new_array,class_num])
                if images_loaded >= LIMIT_PER_CLASS:
                    break
                images_loaded += 1

            except Exeption as e:
                pass
            
            
        
       
```

```python
creat_training_data ()
```

```python
print(len(training_data))
```

```python
import random
random.shuffle(training_data)
```

```python
X = [] #data/img/feature
Y = [] #label

for features,label in training_data:
     X.append(features)
     Y.append(label)

X= np.array(X).reshape(-1, img_size ,img_size, 3)
```

```python
import cv2
import numpy as np
from keras.models import load_model
from PIL import Image
from IPython.display import display, clear_output

# 1. Load your emotion detection model
# Replace 'path_to_your_model.h5' with the actual path to your pre-trained model
new_model = load_model('my_modelfinale1.keras')

# 2. Start video capture using OpenCV
cap = cv2.VideoCapture(0)  # 0 is the default camera

# 3. Face detection using OpenCV's Haar Cascade
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 4. Start capturing frames
def capture_and_process_frame():
    ret, frame = cap.read()  # Capture a frame

    if not ret:
        print("Failed to grab frame.")
        return

    # Convert frame to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Process each face
    for (x, y, w, h) in faces:
        # Draw face rectangle
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # Crop the face region for emotion detection
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        # Resize to match model input size
        final_image = cv2.resize(roi_color, (224, 224))
        final_image = np.expand_dims(final_image, axis=0)  # Add batch dimension
        final_image = final_image / 255.0  # Normalize the image

        # Get predictions from the emotion detection model
        predictions = new_model.predict(final_image)

        # Identify the predicted emotion
        emotion_labels = ["Angry", "Disgust", "fear", "happy", "neutral", "sad", "Surprised"]
        status = emotion_labels[np.argmax(predictions)]

        # Display the emotion label on the frame
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(frame, status, (x, y - 10), font, 0.9, (0, 0, 255), 2)

    # Display the processed frame in Jupyter
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)  # Convert to RGB for Jupyter
    img = Image.fromarray(frame_rgb)  # Convert to PIL image
    clear_output(wait=True)  # Clear previous outputs to avoid stack
    display(img)  # Display the frame

# Capture and display frames in real time
try:
    while True:
        capture_and_process_frame()
        if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to stop
            break
finally:
    cap.release()  # Release the video capture object when done
    cv2.destroyAllWindows()  # Close all OpenCV windows
```
